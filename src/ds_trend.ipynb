```python
import re


def SetDiff(row):
    lst = []
    for l1 in str(row["コアスキル"]).split("#"):
        for l2 in re.split('[,、、・/／（(:：;；のをな]', l1.replace("など", "").replace("等", "").upper()):
            for l3 in l2.split("による"):
                for l4 in l3.split("から"):
                    for l5 in l4.split("のための"):
                        for l6 in l5.split("を用いた"):
                            for l7 in l6.split("での"):
                                for l8 in l7.split("における"):
                                    lst.append(l8.
                                               replace(')', '').
                                               replace('）', '').
                                               replace('の業務経験', '').
                                               replace('の経験', '').
                                               replace('経験', '').
                                               replace('の研究', '').
                                               replace('研究', '').
                                               replace('の開発', '').
                                               replace('開発', '').
                                               replace('の立ち上げ', '').
                                               replace('立ち上げ', '').
                                               replace('GOOGLE BIGQUERY', 'BIGQUERY').
                                               replace('GOOGLE BIG QUERY', 'BIGQUERY').
                                               replace('BIG QUERY', 'BIGQUERY').
                                               replace('GOOGLE CLOUD PLATFORM', 'GCP').
                                               replace('MICROSOFT AZURE', 'AZURE').
                                               replace('データベース', 'SQL').
                                               replace('MACHINE LEARNING', '機械学習').
                                               replace('DX推進', 'DX').
                                               replace('GITLAB', 'GIT').
                                               replace('GITHUB', 'GIT').
                                               replace('DEEP LEARNING', 'ディープラーニング').
                                               replace('機械学習プロジェクト', '機械学習').
                                               replace('テキストマイニング', '自然言語処理').
                                               replace('機械学習モデル構築', 'モデル構築').
                                               replace('AIモデル構築', 'モデル構築').
                                               replace('機械学習エンジニア', '機械学習').
                                               replace('データ前処理', '前処理').
                                               replace('DATA SCIENCE', 'データサイエンス').
                                               replace('R言語', 'R').
                                               replace('EXCEL VBA', 'VBA').
                                               replace('SQL構築', 'SQL').
                                               replace('PYTHONプログラミング', 'PYTHON').
                                               strip()
                                               )
    return lst


#1行ごとにSetDiff()を呼び出す
dfx["スキルリスト"] = dfx.apply(SetDiff, axis=1)
```

```python
dfx["スキルリスト"]
```

```python
word = []
for index, row in dfx.iterrows():
    for l in row["スキルリスト"]:
        word.append(l)
```
